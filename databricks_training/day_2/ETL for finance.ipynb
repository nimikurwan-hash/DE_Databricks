```python
from pyspark.sql.functions import *

df_finance = spark.read.csv('/Volumes/acc_dev/naval/raw/financial_dataset.csv', header=True, inferSchema=True)
```

```python
renamed_column = {}
for columns in df_finance.columns:
    renamed_column[columns] = columns.lower()

#print(renamed_column)
```

```python
df_finance=df_finance.withColumnsRenamed(renamed_column)\
.withColumn("ingestion_date", current_timestamp())\
.withColumn("file_path", lit("/Volumes/acc_dev/naval/raw/financial_dataset.csv"))
```

```python
df_finance.write.mode("overwrite").saveAsTable("acc_dev.nimish.finance")
```
